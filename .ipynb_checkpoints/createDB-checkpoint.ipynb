# Notebook to Create Database

```python
import sqlite3
import pandas as pd
```

Create a database and connect

```python
conn = sqlite3.connect('renewable_spain.db')
c = conn.cursor()
```

Connect to the raw data table

```python
read_raw = pd.read_csv(r'raw_data/energy_dataset.csv')
read_raw.to_sql('energy_dataset', conn, if_exists='append',index = False')
```

Create time_stamp table and add timestamps with index as primary key

```python
c.execute('''
CREATE TABLE time_stamps (
row int PRIMARY KEY,
time text,
) [WITHOUT ROWID];
''')
c.execute('''INSERT INTO time_stamps SELECT time FROM energy_dataset''')
index = 0
for row in c.execute('''SELECT * FROM time_stamps''')
    c.execute('''UPDATE time_stamps SET row = ?''', index)
    index += 1
```
