## Adding Tables to Database

Create time_stamp table and add timestamps with index as primary key this may be unnessary but thought it might be helpful to have a table with just our timestamps if needed. 

```python
c.execute('''
    CREATE TABLE time_stamps (
    row int PRIMARY KEY,
    time text
    );
    ''')
```

Add time stamps with row numbers for the PK

```python
c.execute('''INSERT INTO time_stamps (row, time) SELECT ROW_NUMBER() OVER (ORDER BY time) as row, time FROM energy_dataset''')

for row in c.execute('''SELECT * FROM time_stamps'''):
    print(row)
```

Creating a table to hold the forcasted and actual load data using time as PK

```python
c.execute('''CREATE TABLE load (
    row_key int,
    time text PRIMARY KEY,
    forcasted_load int,
    actual_load int
    );''')
```

```python
c.execute('''
INSERT INTO load (row_key, time, forcasted_load, actual_load) 
SELECT ROW_NUMBER() OVER (ORDER BY total_load_forecast) as row_key,
time,
total_load_forecast,
total_load_actual 
FROM energy_dataset''')
```

```python
for row in c.execute('''SELECT * FROM load LIMIT 5'''):
    print(row)
```

### Visualization of data

```python
import matplotlib.pyplot as plt
```

```python
# Load the data from SQLite database into a DataFrame
conn = sqlite3.connect('renewable_spain.db')
c = conn.cursor()
```

```python
df = pd.read_sql_query("SELECT time, forcasted_load, actual_load FROM load", conn)
```

```python
# Select the desired columns from the table
query = "SELECT time, forcasted_load, actual_load FROM load"

# Load the data into a pandas DataFrame
try:
    df = pd.read_sql_query(query, conn)
    print(df.head())
except Exception as e:
    print("Error: ", e)
finally:
    conn.close()
```

```python
# Convert the 'time' column to a datetime object because SQLite does not allow for time datatype
df['time'] = pd.to_datetime(df['time'])
df = df.sort_values('time', ascending=False)
df = df.tail(50)
# Set the index of the DataFrame to the 'time' column
df.set_index('time', inplace=True)
```

```python
print(len(df))
```

```python
# Plot the data
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df.index, df['forcasted_load'], label='Total Forecasted Load')
ax.plot(df.index, df['actual_load'], label='Total Actual Load')
ax.set_xlabel('Time')
ax.set_ylabel('Load')
ax.set_ylim([0, 50000])
ax.legend()
plt.show()
```

# Database Schema Diagram

```python
import sqlite3

# connect to the database
conn = sqlite3.connect('renewable_spain.db')

# get the schema information
cursor = conn.execute("SELECT sql FROM sqlite_master WHERE type='table'")

# iterate through the tables and print the schema
for table in cursor.fetchall():
    print(table[0])
```

```python
pip install graphviz
```

```python
from graphviz import Digraph

# create a new graph
dot = Digraph(comment='SQLite Database Schema')

# iterate through the tables and add them to the graph
for table in cursor.fetchall():
    # extract the table name from the SQL statement
    table_name = table[0].split()[2]
    # add the table to the graph
    dot.node(table_name, table_name)

# iterate through the tables again and add the relationships
for table in cursor.fetchall():
    # extract the table name from the SQL statement
    table_name = table[0].split()[2]
    # extract the foreign key information from the SQL statement
    foreign_keys = [x.strip().split(' REFERENCES ')[1].split('(')[0] for x in table[0].split('\n') if 'REFERENCES' in x]
    # add the relationships to the graph
    for fk_table in foreign_keys:
        dot.edge(fk_table, table_name)

# save the graph to a file
dot.render('database_schema', view=True)
```

```python
from IPython.display import Image

# Retrieve the schema information
cursor = conn.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = [table[0] for table in cursor.fetchall()]

# Create the ERD diagram
dot = Digraph(comment='Database Schema')
for table in tables:
    cursor = conn.execute(f"PRAGMA table_info('{table}')")
    columns = [column[1] for column in cursor.fetchall()]
    dot.node(table, '\n'.join(columns))

cursor = conn.execute("SELECT * FROM sqlite_master WHERE type='index'")
for index in cursor.fetchall():
    dot.edge(index[2], index[4])

# Display the ERD diagram in the notebook
dot.format = 'png'
Image(dot.render())
```
